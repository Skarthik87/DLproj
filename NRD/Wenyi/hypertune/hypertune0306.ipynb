```python
res = pd.DataFrame()

for job_ind in range(3):
    df = pd.read_csv('output/ht_result0307_'+str(job_ind)+'.csv', 
                     names=['model_name', 'code_embed_dim', 'fc_width', 'md_width', 'lr1', 'lr2', 'dropout',
                            'batchsize', 'embed_file', 'tst_seed', 'n_fold', 'count_cap', 
                            'code_rarecutpoint', 'minor_class_weight', 'auc_mean', 'auc_freeze', 'n_sample'], index_col=None)
    res = pd.concat([res, df])
```
