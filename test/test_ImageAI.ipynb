```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
from PIL import Image
%matplotlib inline
```

```python
from imageai.Detection import ObjectDetection
```

```python
path = '/nfs/turbo/umms-awaljee/wsliu/Data/test/'
data_path = '/nfs/turbo/umms-awaljee/wsliu/Data/test/image_samples/'
model_path = '/nfs/turbo/umms-awaljee/wsliu/Data/test/models/'
```

```python
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath(model_path+'resnet50_coco_best_v2.0.1.h5')
```

```python
detector.loadModel()
```

```python
images = os.listdir(data_path)
```

```python
images
```

```python
for name in images:
    img = Image.open(data_path+name)
    img.thumbnail((1000, 1000), Image.ANTIALIAS)
    img.save(data_path+'resize_'+name)
```

```python
img = 'IMG_0288.JPG'
detections = detector.detectObjectsFromImage(input_image=data_path+'resize_'+img, output_image_path=path+'detection_output/output_'+img)
```

```python
for obj in detections:
    print(obj['name']+':'+obj['percentage_probability'])
```
